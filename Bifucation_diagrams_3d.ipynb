```python
import cmath
import matplotlib.pyplot as plt
import numpy as np

# Define the functions
def fixed_point(f_hat, V, m):
    return f_hat / (V * m)

def stability1(k):
    return -k

def stability_plus(f_hat):
    intermediate = -((5.0 * 10**(-10) * ((1.3782 * (10**9)) * f_hat + (cmath.sqrt(1.8996 * (10**18) * f_hat**2 - 9.4993 * (10**19) * f_hat))))/f_hat)
    value = intermediate.real
    return value

def stability_minus(f_hat):
    intermediate = -((5.0 * 10**(-10) * ((1.3782 * (10**9)) * f_hat - (cmath.sqrt(1.8996 * (10**18) * f_hat**2 - 9.4993 * (10**19) * f_hat))))/f_hat)
    value = intermediate.real
    return value
```

```python
# Params
V = 80000
m = 0.022
k = 0.04

f_hat_values_pos = np.linspace(0.1,100000,1000000)
f_hat_values_neg = np.linspace(-100000, -0.1, 1000000)

#calculating all the stabilities
stable_fixed_points = []
stable_fs = []
unstable_fixed_points = []
unstable_fs = []
for i in range(len(f_hat_values_pos)):
    if stability1(k) < 0 and round(stability_minus(f_hat_values_pos[i]), 200) < 0 and stability_plus(f_hat_values_pos[i]) < 0:
        stable_fixed_points.append(fixed_point(f_hat_values_pos[i], V, m))
        stable_fs.append(f_hat_values_pos[i])
    elif stability1(k) == 0 and stability_minus(f_hat_values_pos[i]) == 0 and stability_plus(f_hat_values_pos[i]) == 0:
        unstable_fixed_points.append(fixed_point(f_hat_values_pos[i], V, m))
        unstable_fs.append(f_hat_values_pos[i])
        
    else:
        unstable_fixed_points.append(fixed_point(f_hat_values_pos[i], V, m))
        unstable_fs.append(f_hat_values_pos[i])
        #print(stability_plus(f_hat_values[i]))
    #print(round(stability_minus(f_hat_values[i]), 3))

stable_fixed_points_neg = []
stable_fs_neg = []
unstable_fixed_points_neg = []
unstable_fs_neg = []
for i in range(len(f_hat_values_neg)):
    if stability1(k) < 0 and stability_minus(f_hat_values_neg[i]) < 0 and stability_plus(f_hat_values_neg[i]) < 0:
        stable_fixed_points_neg.append(fixed_point(f_hat_values_neg[i], V, m))
        stable_fs_neg.append(f_hat_values_neg[i])
    elif stability1(k) == 0 and stability_minus(f_hat_values_neg[i]) == 0 and stability_plus(f_hat_values_neg[i]) == 0:
        unstable_fixed_points_neg.append(fixed_point(f_hat_values_neg[i], V, m))
        unstable_fs_neg.append(f_hat_values_neg[i])

        #print(stability_minus(f_hat_values_neg[i]), stability_plus(f_hat_values_neg[i]))
        
    else:
        unstable_fixed_points_neg.append(fixed_point(f_hat_values_neg[i], V, m))
        unstable_fs_neg.append(f_hat_values_neg[i])
        #print(stability_minus(f_hat_values_neg[i]), stability_plus(f_hat_values_neg[i]))




#plt.rcParams['text.usetex'] = False

plt.figure(figsize=(12,10))
plt.grid()
plt.plot(unstable_fs, unstable_fixed_points, '-o', color='red', label='Unstable')
plt.plot(stable_fs, stable_fixed_points, '-o', color='blue', label='Stable')
plt.plot(unstable_fs_neg, unstable_fixed_points_neg, '-o', color='red')  # Negative unstable
#plt.plot(stable_fs_neg, stable_fixed_points_neg, '-o', color='blue')  # Negative stable
plt.plot(0, 0, 'o', color = 'orange', label = 'Undefined stability')
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.ylabel(r'Fixed point value in $X$', fontsize=14)
plt.xlabel(r'log($\hat{F}$)', fontsize=14)
plt.xscale('symlog')

#plt.yscale('symlog')
plt.legend(fontsize=14)
plt.show()
```
